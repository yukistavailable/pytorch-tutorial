```python
import os
import time
import requests
import json
import nltk
import pickle
from pycocotools.coco import COCO
from build_vocab import Vocabulary
from collections import defaultdict
from torchtext.data.metrics import bleu_score
from sample import generate_sentence, generate_sentence_from_image_url
```

```python
image_path = 'data/images/elephants.png'
encoder_path = 'models/encoder512-2-3000.ckpt'
decoder_path = 'models/decoder512-2-3000.ckpt'
vocab_path = 'data/vocab.pkl'
embed_size = 512
hidden_size = 512
num_layers = 1
```

```python
with open('data/annotations/captions_val2014.json') as f:
    val_text = f.read()
with open('data/annotations/captions_train2014.json') as f:
    train_text = f.read()
val_info = json.loads(val_text)
train_info = json.loads(train_text)
```

```python
val_image_urls = defaultdict(str)
for i in range(len(val_info['images'])):
    val_image_urls[val_info['images'][i]['id']] = val_info['images'][i]['flickr_url']
val_annotatios = defaultdict(list)
for i in range(len(val_info['annotations'])):
    val_annotatios[val_info['annotations'][i]['image_id']].append(val_info['annotations'][i]['caption'])
```

```python
img_id = 471785
image_url = val_image_urls[img_id]
annotations = val_annotatios[img_id]
print(image_url)
print(annotations)
```

```python
sentence = generate_sentence_from_image_url(image_url, encoder_path, decoder_path, vocab_path, embed_size, hidden_size, num_layers)
print(sentence)
```

```python
splitted_annotations = [nltk.word_tokenize(annotation.lower()) for annotation in annotations]
splitted_generated_sentence = nltk.word_tokenize(sentence[8:-6])
```

```python
bleu_score([splitted_generated_sentence], [splitted_annotations])
```
