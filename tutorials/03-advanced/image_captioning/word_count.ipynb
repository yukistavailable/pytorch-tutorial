```python
import json
from collections import defaultdict
```

```python
with open('data/annotations/captions_val2014.json') as f:
    val_text = f.read()
with open('data/annotations/captions_train2014.json') as f:
    train_text = f.read()
val_info = json.loads(val_text)
train_info = json.loads(train_text)
```

```python
total = ''
for a in val_info['annotations']:
    total += a['caption'].lower().replace('.', '')
    total += ' '
```

```python
d = defaultdict(int)
for word in total.split(' '):
    if word != '':
        d[word] += 1
```

```python
with open('val_words_count.json', 'w') as f:
    json.dump(d, f)
```

# sort

```python
with open('words_count.json', 'r') as f:
    generated_words_count = json.load(f)

with open('val_words_count.json', 'r') as f:
    val_words_count = json.load(f)

with open('train_words_count.json', 'r') as f:
    train_words_count = json.load(f)
```

```python
generated_words_count['.'] = 0
generated_words_count = dict(sorted(generated_words_count.items(), key=lambda item: - item[1]))
val_words_count = dict(sorted(val_words_count.items(), key=lambda item: - item[1]))
train_words_count = dict(sorted(train_words_count.items(), key=lambda item: - item[1]))
```

```python
total_generated = sum(generated_words_count.values())
total_train = sum(train_words_count.values())
total_val = sum(val_words_count.values())
```

```python
generated_ratio = defaultdict(float)
train_ratio =  defaultdict(float)
val_ratio = defaultdict(float)
for g in generated_words_count.keys():
    generated_ratio[g] = generated_words_count[g] / total_generated
```

```python
for v in val_words_count.keys():
    val_ratio[v] = val_words_count[v] / total_val
```

```python
abs_difference_gen_val = defaultdict(float)
for v in val_ratio.keys():
    abs_difference_gen_val[v] = abs(val_ratio[v] - generated_ratio[v])
abs_difference_gen_val = dict(sorted(abs_difference_gen_val.items(), key=lambda item: - item[1]))
```
